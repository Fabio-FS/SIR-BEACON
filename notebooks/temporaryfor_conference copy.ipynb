```python
import igraph as ig
import matplotlib.pyplot as plt
import numpy as np

# Create a 7x7 lattice
g = ig.Graph.Lattice(dim=[7, 7], circular=False)

# Add a random attribute between 0 and 1

g.vs["value"] = np.ones(49)
# Map values to colors
from matplotlib.colors import Normalize
norm = Normalize(vmin=min(g.vs["value"]), vmax=max(g.vs["value"]))
g.vs["color"] = [cmap(norm(v)) for v in g.vs["value"]]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
import matplotlib.colors as mcolors

cmap = LinearSegmentedColormap.from_list("pink_green", ["#8E0152", "#BBB", "#276419"])

# Convert to hex strings that igraph understands
g.vs["value"] = np.zeros(49)
vertex_colors = [mcolors.to_hex(cmap(v)) for v in g.vs["value"]]
ig.plot(
    g,
    target=ax,
    layout="grid",
    vertex_color=vertex_colors,
    vertex_size=sizes
)
plt.show()
```

```python
import igraph as ig
import matplotlib.pyplot as plt
import numpy as np

# Create a 7x7 lattice
g = ig.Graph.Lattice(dim=[3, 3], circular=False)

# Add a random attribute between 0 and 1

g.vs["value"] = np.ones(9)
# Map values to colors
g.vs["color"] = [cmap(v) for v in g.vs["value"]]

# Central node index: for 7x7, center is at (3,3) → index 24
central = 4

# Set sizes: all small except central
sizes = [55] * 49
sizes[central] = 55
shapes = ["circle"] * 9
shapes[central] = "square"

fig, ax = plt.subplots(1, 1, figsize=(6, 6))

ig.plot(
    g,
    target=ax,
    layout="grid",
    vertex_color=g.vs["color"],
    vertex_size=sizes,
    vertex_shape=shapes
)

plt.show()
```
