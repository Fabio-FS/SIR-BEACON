```python
def create_parameter_grid(results_list, metric, cmap, nrows, ncols, figsize=(Lx*4.5, Ly*3.5), 
                          contour_values=None, contour_colors="black", 
                          xticks=None, yticks=None, xticklabels=None, yticklabels=None,
                          ticklabel_fontsize=12, vmin=0, vmax=1,
                          colorbar_per_row=True, save_path=None):
    """
    Create a grid of parameter sweep plots with the same metric and formatting
    
    Args:
        results_list: List of results dictionaries
        metric: Metric to plot (e.g., "infections", "r0")
        cmap: Colormap to use for all plots
        nrows, ncols: Grid dimensions
        figsize: Figure size
        contour_values: Contour values for all plots (or None for no contours)
        contour_colors: Contour colors (or None for default white)
        xticks, yticks: Custom tick positions for all plots
        xticklabels, yticklabels: Custom tick labels for all plots
        ticklabel_fontsize: Font size for tick labels
        vmin, vmax: Fixed colorscale limits
        colorbar_per_row: Whether to add one colorbar per row
        save_path: Path to save the figure
        
    Returns:
        Figure with grid of plots
    """
    import matplotlib.pyplot as plt
    import numpy as np
    import jax.numpy as jnp
    from matplotlib.gridspec import GridSpec
    
    # Create figure with a grid that has space for colorbars
    fig = plt.figure(figsize=figsize)
    
    if colorbar_per_row:
        # Create a GridSpec with an extra thin column for colorbars
        gs = GridSpec(nrows, ncols + 1, figure=fig, width_ratios=[1] * ncols + [0.05])
    else:
        # Standard grid without extra column
        gs = GridSpec(nrows, ncols, figure=fig)
    
    # Create axes for each plot
    axes = np.empty((nrows, ncols), dtype=object)
    for i in range(nrows):
        for j in range(ncols):
            axes[i, j] = fig.add_subplot(gs[i, j])
    
    # Create separate axes for colorbars if needed
    cbar_axes = []
    if colorbar_per_row:
        for i in range(nrows):
            cbar_axes.append(fig.add_subplot(gs[i, -1]))
    
    # Create the plots
    fig_idx = 0
    row_images = []  # Store one image per row for colorbars
    
    for i in range(nrows):
        row_images.append(None)  # Placeholder for this row's image reference
        
        for j in range(ncols):
            if fig_idx < len(results_list):
                result = results_list[fig_idx]
                ax = axes[i, j]
                
                # Get the parameter information
                param1_name = result['parameter_names']['param1']
                param2_name = result['parameter_names']['param2']
                
                # Determine if axes should be swapped
                swap_axes = False
                if param1_name == "beta_params" and param2_name != "beta_params":
                    param1_label = "polarization"
                    param2_label = param2_name
                    swap_axes = False
                elif param2_name == "beta_params" and param1_name != "beta_params":
                    param1_label = param1_name
                    param2_label = "polarization"
                    swap_axes = True
                else:
                    param1_label = param1_name
                    param2_label = param2_name
                
                # Get parameter values
                param1_vals = result['parameter_grid']['param1_vals']
                param2_vals = result['parameter_grid']['param2_vals']
                
                # Prepare the data based on the metric
                if metric == "infections":
                    data = np.ones_like(result['r0'])
                    if "S" in result['final_state']:
                        data -= np.array(jnp.sum(result['final_state']["S"], axis=2))
                    if "V" in result['final_state']:
                        data -= np.array(jnp.sum(result['final_state']["V"], axis=2))
                elif metric == "r0":
                    data = np.array(result['r0'])
                elif metric in result['final_state']:
                    data = np.array(jnp.sum(result['final_state'][metric], axis=2))
                else:
                    fig_idx += 1
                    continue
                
                # Handle axis swapping if needed
                if swap_axes:
                    data = data.T
                    plot_param1_vals = param2_vals.T
                    plot_param2_vals = param1_vals.T
                else:
                    plot_param1_vals = param1_vals
                    plot_param2_vals = param2_vals
                
                # Create heatmap
                im = ax.pcolormesh(
                    plot_param1_vals,
                    plot_param2_vals,
                    data,
                    cmap=cmap,
                    vmin=vmin,
                    vmax=vmax
                )
                
                # Store image reference for the first column of each row for colorbars
                if j == 0:
                    row_images[i] = im
                
                # Add contours if specified
                if contour_values and len(contour_values) > 0:
                    try:
                        # Create a meshgrid for contour
                        x_grid = np.linspace(np.min(plot_param1_vals), np.max(plot_param1_vals), data.shape[1])
                        y_grid = np.linspace(np.min(plot_param2_vals), np.max(plot_param2_vals), data.shape[0])
                        X, Y = np.meshgrid(x_grid, y_grid)
                        
                        # Draw contours
                        cs = ax.contour(
                            X, Y, data,
                            levels=contour_values,
                            colors=contour_colors,
                            linewidths=1.0
                        )
                        
                        # Add contour labels
                        ax.clabel(cs, inline=True, fontsize=8, fmt='%g')
                    except Exception as e:
                        print(f"Contour error in subplot ({i},{j}): {e}")
                
                # Set custom ticks and labels
                if xticks is not None:
                    ax.set_xticks(xticks)
                if yticks is not None:
                    ax.set_yticks(yticks)
                if xticklabels is not None:
                    ax.set_xticklabels(xticklabels, fontsize=ticklabel_fontsize)
                if yticklabels is not None:
                    ax.set_yticklabels(yticklabels, fontsize=ticklabel_fontsize)
                
                fig_idx += 1
            else:
                # Hide unused subplots
                axes[i, j].axis('off')
    
    # Add colorbars
    if colorbar_per_row:
        # One colorbar per row using the dedicated axes
        for i, im in enumerate(row_images):
            if im is not None:
                cbar = fig.colorbar(im, cax=cbar_axes[i])
                
                # Set colorbar label based on metric (only for middle row)
                if i == nrows // 2:
                    if metric == "infections":
                        cbar.set_label("Fraction Infected", fontsize=ticklabel_fontsize)
                    elif metric == "r0":
                        cbar.set_label("R0 Value", fontsize=ticklabel_fontsize)
                    else:
                        cbar.set_label(metric, fontsize=ticklabel_fontsize)
                
                # Set tick label fontsize
                cbar.ax.tick_params(labelsize=ticklabel_fontsize)
    else:
        # One shared colorbar for all plots
        cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
        if row_images[0] is not None:
            cbar = fig.colorbar(row_images[0], cax=cbar_ax)
            
            # Set colorbar label based on metric
            if metric == "infections":
                cbar.set_label("Fraction Infected", fontsize=ticklabel_fontsize)
            elif metric == "r0":
                cbar.set_label("R0 Value", fontsize=ticklabel_fontsize)
            else:
                cbar.set_label(metric, fontsize=ticklabel_fontsize)
            
            # Set tick label fontsize
            cbar.ax.tick_params(labelsize=ticklabel_fontsize)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=1200, bbox_inches='tight')
    
    return fig, axes, cbar_axes
```

```python
combined_fig, plot_axes, cbar_axes = create_parameter_grid(
    results_list=[results_MASKS_5,results_MASKS_25, results_MASKS_100, results_MASKS_500, results_TESTS_5, results_TESTS_25, results_TESTS_100, results_TESTS_500, results_VACCINATION_5, results_VACCINATION_25, results_VACCINATION_100, results_VACCINATION_500],
    metric="infections",
    cmap=my_hot_r,
    nrows=3, 
    ncols=4,
    contour_values=[0.25, 0.5, 0.75],
    contour_colors="black",
    xticks=[0, 0.5, 1],
    yticks=[-4, 0, 4],
    xticklabels=["0",  "0.5",  "1"],
    yticklabels=["-4", "0", "4"]
)

plot_axes[0, 0].set_ylabel("SIRM\n\nHomophily", fontsize=14)
plot_axes[1, 0].set_ylabel("SIRT\n\nHomophily", fontsize=14)
plot_axes[2, 0].set_ylabel("SIRV\n\nHomophily", fontsize=14)

plot_axes[2, 0].set_xlabel("Polarization", fontsize=14)
plot_axes[2, 1].set_xlabel("Polarization", fontsize=14)
plot_axes[2, 2].set_xlabel("Polarization", fontsize=14)
plot_axes[2, 3].set_xlabel("Polarization", fontsize=14)


cbar_axes[0].set_ylabel("Fraction of Infected", fontsize=14)
cbar_axes[0].set_yticks([0, 0.5, 1])
cbar_axes[1].set_ylabel("Fraction of Infected", fontsize=14)
cbar_axes[1].set_yticks([0, 0.5, 1])
cbar_axes[2].set_ylabel("Fraction of Infected", fontsize=14)
cbar_axes[2].set_yticks([0, 0.5, 1])


plot_axes[0, 0].set_title("N = 5", fontsize=14)
plot_axes[0, 1].set_title("N = 25", fontsize=14)
plot_axes[0, 2].set_title("N = 100", fontsize=14)
plot_axes[0, 3].set_title("N = 500", fontsize=14)


combined_fig.savefig("figures/SI/Dependency_on_NCOMPARTMENTS.pdf", dpi=150, bbox_inches='tight')
```
