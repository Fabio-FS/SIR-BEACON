```python
%load_ext autoreload
%autoreload 2
# example.py
import numpy as np
from src.utils.visualization_restructuring.factory import PlotFactory
import matplotlib.pyplot as plt

# Test HeatmapBuilder
def test_heatmap():
    # Generate sample data
    x = np.linspace(-3, 3, 100)
    y = np.linspace(-3, 3, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.sin(X) * np.cos(Y)
    
    # Create heatmap using builder
    builder = PlotFactory.create_builder('heatmap')
    fig, ax = builder.set_data(Z, x, y) \
                    .set_colormap('viridis') \
                    .set_colorbar(True, 'Value') \
                    .set_contours(True) \
                    .set_title('Heatmap Example') \
                    .set_labels('X-axis', 'Y-axis') \
                    .build()
    
    plt.show()

# Test TimeSeriesBuilder
def test_timeseries():
    # Generate sample data
    t = np.linspace(0, 10, 100)
    y1 = np.sin(t)
    y2 = np.cos(t)
    
    # Create time series using builder
    builder = PlotFactory.create_builder('timeseries')
    fig, ax = builder.add_series(y1, t, 'sin(t)') \
                    .add_series(y2, t, 'cos(t)') \
                    .set_colors(['blue', 'red']) \
                    .set_title('Time Series Example') \
                    .set_labels('Time', 'Value') \
                    .set_grid(True) \
                    .build()
    
    plt.show()

if __name__ == "__main__":
    test_heatmap()
    test_timeseries()
```
